```python
from numpy import genfromtxt, savetxt

import numpy as np

data_set = np.array(genfromtxt('../RESOURCE/data_set.csv',delimiter = ",", usecols = range(1,21), dtype="str"))

# Column containing 'Yes', 'No' are
# Dependents
# Partner
# PhoneService
# MultipleLines
# OnlineSecurity, 
# Onlinebackup, 
# DeviceProtection, 
# TechSupport, 
# StreamingTV, 
# StreamingMovies
# PaperlessBilling
# Churn
data_set = np.where(data_set == "Yes", 1, data_set)
data_set = np.where(data_set=="No", 0, data_set)

# Column containing 'Male', 'Female' are
# Gender
data_set = np.where(data_set=="Male", 1, data_set)
data_set = np.where(data_set=="Female", 0, data_set)

# Column containing 'No phone service' are
# MultipleLines
data_set = np.where(data_set=="No phone service", 2, data_set)

# Column containing 'DSL, Fibre optic' are
# Internet service
data_set = np.where(data_set=="DSL", 1, data_set)
data_set = np.where(data_set=="Fiber optic", 2, data_set)

# Column containing 'No internet service' are
# OnlineSecurity, 
# OnlineBackup, 
# DeviceProtection, 
# TechSupport, 
# StreamingTV, 
# StreamingMovies
data_set = np.where(data_set=="No internet service", 1, data_set)

# Column containing 'Month-to-month','One year','Two year' are
# Contract
data_set = np.where(data_set=="Month-to-month", 0, data_set)
data_set = np.where(data_set=="One year", 1, data_set)
data_set = np.where(data_set=="Two year", 2, data_set)

# Column containg 'Electronic check','Mailed check', Bank transfer (automatic)
# PaymentMethod
data_set = np.where(data_set=="Electronic check", 0, data_set)
data_set = np.where(data_set=="Mailed check", 1, data_set)
data_set = np.where(data_set=="Bank transfer (automatic)", 2, data_set)
data_set = np.where(data_set=="Credit card (automatic)", 3, data_set)

# Writting the numpy array to a csv file
#savetxt('../RESOURCE/processed1_data.csv',data_set,delimiter = ",", fmt='%s')
```
